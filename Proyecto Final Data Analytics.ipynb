```python
precio_canasta_basica=canasta_basica[['AÑO', 'PRECIO CANASTA BASICA']]
precio_canasta_basica.head()
```

```python
precio_canasta_basica_mes=precio_canasta_basica.sort_values(by='AÑO').groupby('AÑO').sum()  #Se convirtio de esta manera el año a indice
precio_canasta_basica_mes.head()
```

```python
precio_canasta_basica_mes_re = precio_canasta_basica_mes['PRECIO CANASTA BASICA'].resample('MS').sum()  # Month Start
precio_canasta_basica_mes_re
```

```python
fig = plt.figure(facecolor='w')
precio_canasta_basica_mes_re.plot(figsize=[15, 6], marker='o', markersize=10, markerfacecolor='b');
```

```python
import statsmodels.api as sm 
from pylab import rcParams

rcParams['figure.figsize'] = 15, 10

decomposition = sm.tsa.seasonal_decompose(precio_canasta_basica_mes_re, model='additive')
fig = decomposition.plot()
fig.show();
```

```python
from statsmodels.tsa.stattools import adfuller
```

H0 = Nuestra serie de tiempo NO es estacionaria.

H1 = Nuestra serie de tiempo Sí es estacionaria.

```python
test_adf = adfuller(precio_canasta_basica_mes_re)

test_adf
```

En este caso la hipotesis nula no se rechaza porque la serie de tiempo NO es estacionaria.

Entonces para poder usar un modelo de autoregresion debemos hacer una diferenciacion en nuestros datos y volver a realizar el adfuller



```python
precio_canasta_basica_mes['difference_1'] = precio_canasta_basica_mes['PRECIO CANASTA BASICA'] - precio_canasta_basica_mes['PRECIO CANASTA BASICA'].shift(1)
```

```python
precio_canasta_basica_mes.head()
```

```python
test_adf = adfuller(precio_canasta_basica_mes['difference_1'].dropna())

test_adf
```

Vamos a aplicar modelo SARIMAX

```python
import itertools
```

```python
p = d = q = range(0, 2)
```

```python
p, d , q
```

```python
pdq = list(itertools.product(p, d, q))
pdq
```

```python
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
seasonal_pdq
```

```python
model_param = list()
aic = list()

for param in pdq:
  for param_estacional in seasonal_pdq:
    try:
      model = sm.tsa.statespace.SARIMAX(
          precio_canasta_basica_mes['difference_1'].dropna(), # Serie de tiempo
          order=param, # pdq
          seasonal_order=param_estacional, # PDQ, 12
          enforce_stationarity=False
      )

      results = model.fit()
      model_param.append((param, param_estacional))
      aic.append(results.aic)
      print(f'SARIMAX {param} x {param_estacional}, 12 - AIC: {results.aic}')
    except:
      print('Algo salió mal.')
      continue
```

```python
df_model = pd.DataFrame(zip(model_param, aic), columns=['parameters', 'aic'])
df_model.head()
```

```python
df_model.sort_values('aic').head()
```

```python
model = sm.tsa.statespace.SARIMAX(
    precio_canasta_basica_mes['difference_1'].dropna(),
    order=(1, 1, 1),
    seasonal_order=(0, 1, 1, 12),
    enforce_stationarity=False
)

results = model.fit()
```

```python
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
pred = results.get_prediction(start=pd.to_datetime('2021-01-01'))
pred_conf_int = pred.conf_int()
pred_conf_int
```

```python
ax = precio_canasta_basica_mes['difference_1'].dropna()['2015':].plot(label='Datos Observados')

pred.predicted_mean.plot(ax=ax, label='Predicciones', alpha=0.7, figsize=(15, 10))

ax.fill_between(pred_conf_int.index, 
                pred_conf_int.iloc[:, 0], 
                pred_conf_int.iloc[:, 1], 
                color='g', 
                alpha=0.3)

plt.legend()
plt.show();
```

```python
prediccion_futura = results.get_forecast(steps=3)
pred_ic = prediccion_futura.conf_int()
ax = precio_canasta_basica_mes['difference_1'].dropna()['2017':].dropna().plot(label='Pasado')

prediccion_futura.predicted_mean.plot(ax=ax, label='Pronóstico', alpha=0.7)

ax.fill_between(pred_ic.index,
                pred_ic.iloc[:, 0], 
                pred_ic.iloc[:, 1], 
                color='g', 
                alpha=0.3)


plt.legend()
plt.show()
```
